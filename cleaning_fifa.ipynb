### Fixing dataset headers

```python
# To simplify working with the dataframe, we need to format the headers to lowercase and remove any non-letter symbols from them

fifa.columns = fifa.columns.str.lower().str.replace(' ','_')
```

```python
def headers_cleaner (title):
    return re.sub(r'[^a-zA-Z]', '', title)
```

```python
# applying the function to the dataframe columns

fifa.columns = [headers_cleaner(col) for col in fifa.columns]
```

### Weight and Height columns

```python
#cheking data type for the age, height and weight columns
fifa.dtypes[['height','weight']]
```

```python
# lets have a look at the data format
fifa[['height','weight']]
```

```python
# For analysis purposes, we need to convert "lbs" to kilograms and change the data type to integer

def height_weight_convert(x):
    if x.endswith('lbs') == True:
        x = x.replace('lbs','')
        a = round(int(x)*.453592)
        return a
    
    else:
        x = x.replace("\'"," ").replace("\"","").split(" ")
        b = (int(x[0])*12 + int(x[1]))*2.54
        return b
```

```python
fifa['height'] = fifa['height'].map(height_weight_convert)
fifa['weight'] = fifa['weight'].map(height_weight_convert)
```

```python
fifa[['height', 'weight']]
```

### Value, Wage, Release close

```python
fifa[['value','wage','releaseclause']]
```

```python
fifa['value_duplicated'] = fifa['value']
```

```python
def remove_numbers(x):
    a = re.sub(r'\d+', '', str(x))
    a = a.lstrip('€').replace('.','')
    return a
```

```python
fifa['value_duplicated'] = fifa['value_duplicated'].apply(remove_numbers)
```

```python
fifa['value_duplicated'].value_counts()
```

```python
def euro_convert(x):
    x = x.lstrip('€')
    
    if x.endswith('K') == True:
        a = float(x.rstrip('K'))

    elif x.endswith('M') == True:
        a = float(x.rstrip('M'))*1000
        
    else:
        a = float(x)/1000
        
    return a
```

```python
fifa['value2'] = fifa['value'].apply(euro_convert)
```

```python
fifa[['value','value2']]
```

```python
fifa['value'] = fifa['value2']
```

```python
fifa = fifa.drop(['value2','value_duplicated'], axis = 1)
```

```python
fifa.columns
```

```python
fifa['wage'] = fifa['wage'].apply(euro_convert)
fifa['releaseclause'] = fifa['releaseclause'].apply(euro_convert)
```

```python
fifa[['value','wage','releaseclause']]
```
